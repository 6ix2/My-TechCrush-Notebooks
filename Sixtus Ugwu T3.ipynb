```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

sns.set() #seaborn default setting
```

```python
Orders_df = pd.read_csv("../content/drive/MyDrive/orders.csv")
Orders_df.head()
```

```python
Order_details_df = pd.read_csv("../content/drive/MyDrive/order_details.csv")
Order_details_df.head()
```

```python
Pizzas_df = pd.read_csv("../content/drive/MyDrive/pizzas.csv")
Pizzas_df.head()
```

```python
Pizza_types_df = pd.read_csv("../content/drive/MyDrive/pizza_types.csv", encoding='latin-1')
Pizza_types_df.head()
```

```python
merged_df1 = pd.merge(Orders_df,Order_details_df, on = 'order_id')
merged_df1
```

```python
merged_df2 = pd.merge(Pizzas_df,Pizza_types_df, on = 'pizza_type_id')
merged_df2
```

```python
merged_df3 = pd.merge(merged_df1,merged_df2, on = 'pizza_id')
merged_df3
```

```python
merged_df3.head()
```

```python
merged_df3.info()
```

```python

merged_df3.isna().sum() #using the null and sum fxns to check if there's null value. Data is clean.
```

#what is the Total Revenue/sales?
Total revenue = the sum of total price of pizzas * total quantity of pizzas

```python
#Q1

total_revenue = (merged_df3['quantity'] * merged_df3['price']).sum()
print(f"The Total Revenue made is: ${total_revenue:.2f}") #f string formatted and to 2 decimal place using .2f
```

```python
#Q2 Total Quantity Sold is the sum of all the pizza sales completed

total_quantity_sold = merged_df3['quantity'].sum()
print(f"The Total Quantity Sold is: ${total_quantity_sold}")
```

```python
#Q3 The Total orders was got using the unique fxn from the values of order_id repping total order

total_orders = merged_df3['order_id'].nunique()
print(f"The Total Orders: {total_orders}")
```

```python
#Q4 To find how many pizza types they sell

num_pizza_types = merged_df3['name'].nunique()
print(f"The number of Pizza types are: {num_pizza_types}") #use the unique fxn from the pizza names to extract the diff pizzas thy sell
```

```python
#Q5 To determine average price of pizza using the mean fxn to 2 decimal place

avg_price = merged_df3['price'].mean()
print(f"The Average Price of Pizzas is: ${avg_price:.2f}")
```

```python
#Q6 Find the peak hours of sales

merged_df3['hour'] = pd.to_datetime(merged_df3['time']).dt.hour
peak_hours = merged_df3['hour'].value_counts().sort_values(ascending=False).head()
print("Peak Hours of Sales:")
print(peak_hours)

#changed to datetime and used value_counts fxn to extract peak sales time in descending order
```

```python
#Q7a Total sales made on each day of the week
#changed to datetime down to day of the week and used the groupby and sum fxn

merged_df3['date'] = pd.to_datetime(merged_df3['date'])
merged_df3['day_of_week'] = merged_df3['date'].dt.day_name()
sales_by_day = merged_df3.groupby('day_of_week')['price'].sum().reset_index()
print(sales_by_day)
```

```python
#Q7b Using the sort_values fxn to extract the busiest sales day

busiest_day = sales_by_day.sort_values(by='price', ascending=False).head(1)
print(f"The busiest day is [TGIF] {busiest_day['day_of_week'].values[0]} with total sales of ${busiest_day['price'].values[0]:.2f}")
```

```python
#Q8 Top 5 bestselling pizzas using .head fxn

top_pizzas = merged_df3.groupby('name')['quantity'].sum().reset_index()

top_pizzas = top_pizzas.sort_values(by='quantity', ascending=False).head(5)

print(top_pizzas)
```

```python
#Q9 Sales made in each day of the week
# just like we did in no7a, change the datetime to month name, groupby and sum with ur fxns

merged_df3['date'] = pd.to_datetime(merged_df3['date'])
merged_df3['month'] = merged_df3['date'].dt.month_name()
sales_by_month = merged_df3.groupby('month')['price'].sum().reset_index()
print(sales_by_month)
```

```python
#Q9b Noticeable trend

plt.figure(figsize=(10,6))
sns.barplot(x='month', y='price', data=sales_by_month, palette = 'pastel')
plt.title('DISTRIBUTION OF MONTHLY SALES TREND')
plt.xlabel('MONTHS')
plt.ylabel('SALES')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('DISTRIBUTION OF MONTHLY SALES TREND.png')

plt.show()
```

From the chart you can make out that there have been consistent sales all through the year but some months with of favorable seasons and holidays like summer, Christmas have significant sales.

```python
merged_df3['date'] = pd.to_datetime(merged_df3['date'])
merged_df3['month'] = merged_df3['date'].dt.month
seasonal_sales = merged_df3.groupby('month')['price'].sum().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(x='month', y='price', data=seasonal_sales, color ='blue')
plt.title('DISTRIBUTION OF MONTHLY SALES TREND')
plt.xlabel('MONTH')
plt.ylabel('SALES')
plt.savefig('DISTRIBUTION OF MONTHLY SALES TREND.png')

plt.show()
```

From the trend, holidays, summer seasons and sports activities or carnivals have shown increased sales

```python
#Q10 To find out under performing pizzas

pizza_sales = merged_df3.groupby('name')['quantity'].sum().reset_index()
underperforming_pizzas = pizza_sales.sort_values(by='quantity', ascending=True).head()
print(underperforming_pizzas)
```
